```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder
import warnings
warnings.filterwarnings('ignore')

print("✅ All libraries imported successfully!")
```

```python
# Create the toy dataset as shown in the assignment
toy_data = {
    'TID': ['T1', 'T2', 'T3', 'T4', 'T5'],
    'Items': [
        ['Milk', 'Bread', 'Butter'],
        ['Milk', 'Bread'],
        ['Bread', 'Butter'],
        ['Milk', 'Bread', 'Eggs'],
        ['Milk', 'Eggs']
    ]
}

# Display the transactions
print("📦 Toy Dataset Transactions:")
print("="*50)
for tid, items in zip(toy_data['TID'], toy_data['Items']):
    print(f"{tid}: {', '.join(items)}")
print("="*50)
```

```python
# Convert transactions to one-hot encoded format
te = TransactionEncoder()
te_ary = te.fit(toy_data['Items']).transform(toy_data['Items'])
toy_df = pd.DataFrame(te_ary, columns=te.columns_)

# Add TID column
toy_df.insert(0, 'TID', toy_data['TID'])

print("\n📋 One-Hot Encoded Transaction Matrix:")
print(toy_df)
print("\n✅ True means the item is present in that transaction")
```

```python
# Manual calculations for Support
total_transactions = len(toy_data['TID'])
print(f"Total Transactions: {total_transactions}\n")

# Count item occurrences
items_count = {
    'Bread': 4,
    'Milk': 4,
    'Butter': 2,
    'Eggs': 2
}

# Calculate Support for individual items
print("📈 Support for Individual Items:")
print("="*50)
support_dict = {}
for item, count in items_count.items():
    support = count / total_transactions
    support_dict[item] = support
    print(f"Support({item}) = {count}/{total_transactions} = {support:.2f}")

print("\n" + "="*50)

# Calculate Support for itemsets
itemsets_count = {
    ('Milk', 'Bread'): 3,
    ('Bread', 'Butter'): 2,
    ('Milk', 'Eggs'): 2,
    ('Butter', 'Bread'): 2,
    ('Eggs', 'Milk'): 2
}

print("\n📊 Support for Itemsets (2-item combinations):")
print("="*50)
itemset_support = {}
for itemset, count in itemsets_count.items():
    support = count / total_transactions
    itemset_support[itemset] = support
    print(f"Support{itemset} = {count}/{total_transactions} = {support:.2f}")

print("\n" + "="*50)
```

```python
# Calculate Confidence for key rules
print("\n🔍 Confidence Calculations:")
print("="*50)

# Rule: Butter → Bread
conf_butter_bread = itemset_support[('Butter', 'Bread')] / support_dict['Butter']
print(f"Confidence(Butter → Bread) = {itemset_support[('Butter', 'Bread')]:.2f} / {support_dict['Butter']:.2f} = {conf_butter_bread:.2f}")

# Rule: Eggs → Milk
conf_eggs_milk = itemset_support[('Eggs', 'Milk')] / support_dict['Eggs']
print(f"Confidence(Eggs → Milk) = {itemset_support[('Eggs', 'Milk')]:.2f} / {support_dict['Eggs']:.2f} = {conf_eggs_milk:.2f}")

# Rule: Milk → Bread
conf_milk_bread = itemset_support[('Milk', 'Bread')] / support_dict['Milk']
print(f"Confidence(Milk → Bread) = {itemset_support[('Milk', 'Bread')]:.2f} / {support_dict['Milk']:.2f} = {conf_milk_bread:.2f}")

# Rule: Bread → Milk
conf_bread_milk = itemset_support[('Milk', 'Bread')] / support_dict['Bread']
print(f"Confidence(Bread → Milk) = {itemset_support[('Milk', 'Bread')]:.2f} / {support_dict['Bread']:.2f} = {conf_bread_milk:.2f}")

print("\n" + "="*50)
```

```python
# Calculate Lift
print("\n📈 Lift Calculations:")
print("="*50)

# Lift: Butter → Bread
lift_butter_bread = conf_butter_bread / support_dict['Bread']
print(f"Lift(Butter → Bread) = {conf_butter_bread:.2f} / {support_dict['Bread']:.2f} = {lift_butter_bread:.4f}")

# Lift: Eggs → Milk
lift_eggs_milk = conf_eggs_milk / support_dict['Milk']
print(f"Lift(Eggs → Milk) = {conf_eggs_milk:.2f} / {support_dict['Milk']:.2f} = {lift_eggs_milk:.4f}")

# Lift: Milk → Bread
lift_milk_bread = conf_milk_bread / support_dict['Bread']
print(f"Lift(Milk → Bread) = {conf_milk_bread:.2f} / {support_dict['Bread']:.2f} = {lift_milk_bread:.4f}")

# Lift: Bread → Milk
lift_bread_milk = conf_bread_milk / support_dict['Milk']
print(f"Lift(Bread → Milk) = {conf_bread_milk:.2f} / {support_dict['Milk']:.2f} = {lift_bread_milk:.4f}")

print("\n" + "="*50)
print("\n💡 Interpretation:")
print("   • Lift > 1: Positive correlation")
print("   • Lift = 1: No correlation")
print("   • Lift < 1: Negative correlation")
```

```python
# Create summary table
summary_data = [
    ['Butter', 'Bread', 0.4, 1.00, 1.2500],
    ['Eggs', 'Milk', 0.4, 1.00, 1.2500],
    ['Milk', 'Bread', 0.6, 0.75, 0.9375],
    ['Bread', 'Milk', 0.6, 0.75, 0.9375]
]

summary_df = pd.DataFrame(summary_data,
                         columns=['antecedents', 'consequents', 'support', 'confidence', 'lift'])

print("\n📊 Complete Summary Table:")
print("="*70)
print(summary_df.to_string(index=False))
print("="*70)
```

```python
# Prepare data for Apriori (remove TID column)
toy_df_apriori = toy_df.drop('TID', axis=1)

# Run Apriori algorithm
print("🔍 Running Apriori Algorithm...\n")
frequent_itemsets = apriori(toy_df_apriori, min_support=0.4, use_colnames=True)

print("📊 Frequent Itemsets (min_support = 0.4):")
print("="*70)
print(frequent_itemsets)
print("="*70)
```

```python
# Generate association rules
rules_apriori = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

# Sort by confidence
rules_apriori = rules_apriori.sort_values('confidence', ascending=False)

print("\n🎯 Association Rules (min_confidence = 0.6):")
print("="*70)
print(rules_apriori[['antecedents', 'consequents', 'support', 'confidence', 'lift']].to_string(index=False))
print("="*70)
```

```python
# Run FP-Growth algorithm
print("🌳 Running FP-Growth Algorithm...\n")
frequent_itemsets_fp = fpgrowth(toy_df_apriori, min_support=0.4, use_colnames=True)

print("📊 Frequent Itemsets using FP-Growth (min_support = 0.4):")
print("="*70)
print(frequent_itemsets_fp)
print("="*70)

# Generate rules
rules_fp = association_rules(frequent_itemsets_fp, metric="confidence", min_threshold=0.6)
rules_fp = rules_fp.sort_values('confidence', ascending=False)

print("\n🎯 Association Rules using FP-Growth:")
print("="*70)
print(rules_fp[['antecedents', 'consequents', 'support', 'confidence', 'lift']].to_string(index=False))
print("="*70)
```

```python
# Load the bread basket dataset
print("📂 Loading bread_basket.csv...\n")

# Read the CSV file
bread_basket = pd.read_csv('bread_basket.csv')

print("✅ Dataset loaded successfully!")
print(f"\n📊 Dataset Shape: {bread_basket.shape}")
print(f"   - Total rows: {bread_basket.shape[0]:,}")
print(f"   - Columns: {bread_basket.shape[1]}")

print("\n🔍 First 10 rows:")
print("="*70)
print(bread_basket.head(10))
print("="*70)
```

```python
# Use only the first 200 transactions as specified
print("\n⚙️ Processing first 200 transactions...\n")

# Group by transaction to get list of items per transaction
bread_basket_subset = bread_basket.head(200)
transactions = bread_basket_subset.groupby('transaction')['item'].apply(list).values.tolist()

print(f"✅ Total unique transactions: {len(transactions)}")
print(f"\n📦 Sample transactions:")
print("="*70)
for i, trans in enumerate(transactions[:5], 1):
    print(f"Transaction {i}: {trans}")
print("="*70)
```

```python
# Convert transactions to one-hot encoded format
te_basket = TransactionEncoder()
te_basket_ary = te_basket.fit(transactions).transform(transactions)
basket_df = pd.DataFrame(te_basket_ary, columns=te_basket.columns_)

print("\n✅ One-hot encoding completed!")
print(f"\n📊 Encoded DataFrame Shape: {basket_df.shape}")
print(f"   - Transactions: {basket_df.shape[0]}")
print(f"   - Unique Items: {basket_df.shape[1]}")

print("\n🔍 First 5 rows of encoded data:")
print("="*70)
print(basket_df.head())
print("="*70)
```

```python
# Run FP-Growth algorithm
print("\n🌳 Running FP-Growth on bread_basket data...\n")
print("⚙️ Parameters: min_support = 0.01 (1% of transactions)\n")

freq_fp = fpgrowth(basket_df, min_support=0.01, use_colnames=True)

# Sort by support
freq_fp = freq_fp.sort_values('support', ascending=False)

print("✅ FP-Growth completed!")
print(f"\n📊 Found {len(freq_fp)} frequent itemsets\n")

print("🔝 Top 20 Frequent Itemsets:")
print("="*70)
print(freq_fp.head(20).to_string(index=False))
print("="*70)
```

```python
# Generate association rules
print("\n🎯 Generating Association Rules...\n")
print("⚙️ Parameters: metric = 'confidence', min_threshold = 0.3\n")

rules_basket = association_rules(freq_fp, metric="confidence", min_threshold=0.3)

# Sort by lift to find most interesting rules
rules_basket = rules_basket.sort_values('lift', ascending=False)

print(f"✅ Generated {len(rules_basket)} association rules\n")

print("🏆 Top 20 Rules by Lift (most interesting associations):")
print("="*100)
print(rules_basket[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(20).to_string(index=False))
print("="*100)
```

```python
# Select top rules for visualization
top_rules = rules_basket[(rules_basket['lift'] > 1) & (rules_basket['confidence'] > 0.4)].head(15)

print("\n🎨 Creating Network Visualization...\n")

# Create directed graph
G = nx.DiGraph()

# Add edges
for idx, row in top_rules.iterrows():
    antecedent = ', '.join(list(row['antecedents']))
    consequent = ', '.join(list(row['consequents']))
    G.add_edge(antecedent, consequent, weight=row['lift'])

# Create visualization
plt.figure(figsize=(16, 12))
pos = nx.spring_layout(G, k=2, iterations=50, seed=42)

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=3000, node_color='lightblue',
                       alpha=0.9, edgecolors='navy', linewidths=2)

# Draw edges
edges = G.edges()
weights = [G[u][v]['weight'] for u, v in edges]
nx.draw_networkx_edges(G, pos, width=[w*2 for w in weights],
                       alpha=0.6, edge_color='gray',
                       arrowsize=20, arrowstyle='->',
                       connectionstyle='arc3,rad=0.1')

# Draw labels
nx.draw_networkx_labels(G, pos, font_size=9, font_weight='bold')

# Add edge labels
edge_labels = {(u, v): f"Lift: {G[u][v]['weight']:.2f}" for u, v in G.edges()}
nx.draw_networkx_edge_labels(G, pos, edge_labels, font_size=7)

plt.title('Association Rules Network Visualization\n(Edge thickness = Lift strength)',
          fontsize=16, fontweight='bold', pad=20)
plt.axis('off')
plt.tight_layout()
plt.show()

print("\n✅ Visualization complete!")
```

Q1 & Q2: What does "True" mean in this matrix?
Answer:
In the one-hot encoded matrix, "True" means that the specific item (column) is present in that particular transaction (row). For example, if row T1 has "True" under the "Milk" column, it means transaction T1 contains Milk.
Conversely, "False" means the item is not present in that transaction. This binary representation is essential for association rule mining algorithms to efficiently calculate support, confidence, and other metrics.

Q5: Do Apriori and FP-Growth produce the same rules?
Answer:
Yes, Apriori and FP-Growth produce exactly the same rules on this toy dataset. Both algorithms are designed to find all frequent itemsets that meet the minimum support threshold. The key difference is HOW they find these itemsets, not WHAT they find:

Apriori: Uses breadth-first search with candidate generation
FP-Growth: Uses depth-first search with an FP-tree structure

On large datasets, FP-Growth is typically 10-100x faster because it only scans the database twice and uses a compressed tree structure.
